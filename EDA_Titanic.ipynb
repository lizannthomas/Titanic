**UNIVARIATE AND BIVARIATE ANALYSIS**

**Univariate Analysis:** Data that just contains one variable is referred to as univariate data. Since there is only one variable that varies, univariate data analysis is the most straightforward type of analysis. The analysis's primary goal is to explain the data and identify any patterns in it; it does not deal with causes or correlations. Height is an example of a univariate data.

Let's say that a class of seven kids had their heights measured; the sole variable in this case was height, and it had nothing to do with causes or relationships. Drawing inferences from central tendency measures (mean, median, and mode), dispersion or spread of data (range, minimum, maximum, quartiles, variance, and standard deviation), frequency distribution tables, histograms, pie charts, frequency polygons, and bar charts can all be used to describe patterns found in this type of data.

**Bivariate Analysis:** the kind of data that has two distinct variables. The analysis of this kind of data focuses on relations and causes, and it seeks to understand the logical connection between the two variables. The summertime temperature and ice cream sales are two examples of bivariate data.

Assume that the two variables in a bivariate data are temperature and ice cream sales . In this case, the table makes the link between temperature and sales clear. The two variables are related because as the temperature rises, so do the sales. Accordingly, bivariate data analysis comprises contrasts, connections, causes, and explanations. To better comprehend the data, these variables are frequently shown on the graph's X and Y axes. One of these variables is independent, and the other is dependent.

**Countplot:**

The countplot is used to represent the occurrence(counts) of the observation present in the categorical variable.
It uses the concept of a bar chart for the visual depiction.



```python
sns.set_style('whitegrid')
sns.countplot(x='survived',data=df)
```

From this plot we can understand that about 550 people died and 350 people survived in the Titsnic incident.

```python
sns.countplot(x='survived',hue='sex',data=df)
```

From this plot among the 550 people who are dead 460 are male and 90 are female and among the survived people almost 110 are male and 240 are female. That is, more men have died and more women have survived.

```python
sns.countplot(x='survived',hue='pclass',data=df)
```

Among the people who are dead, about 80 are from class 1, 95 from class 2 and 335 are from class 3. Also among the people who survived about 140 are from class 1, 90 are from class 2 and 120 are from class 3. That is, more people from class 3 has died and more people from class 1 has survived

```python
sns.countplot(x='survived',hue='embarked',data=df)
```

More than 70% of the people boarded from Southampton, under 20% boarded from Cherbourg and the rest boarded from Queenstown. So more people who are dead and survived are from Southampton itself

```python
sns.countplot(x='survived',hue='parch',data=df)
```

**MULTIVARIATE ANALYSIS**

Multivariate data refers to data that has three or more variables. For instance, if an online marketer wanted to compare the popularity of four ads, they might analyse the click rates for men and women and then look at the correlations between the variables.

It is comparable to bivariate but has more dependent variables than that. The methods used to analyse this data depend on the objectives to be met. Regression analysis, path analysis, factor analysis, and multivariate analysis of variance are a few of the methodologies.

```python
df.corr(method='pearson')
```

**Heatmap:**

A depiction of rectangular data presented as a matrix with colours is called a heatmap. It accepts a 2D dataset as a parameter. It is possible to force that dataset into an ndarray. This kind of data visualisation is excellent since it can display the relationship between variables, including time.

**DATA CLEANING**

```python
sns.boxplot(x='pclass',y='age',data=df)
```

From this boxplot we can find the medians of age with respect to pclass of the people in Titanic. Median of class 1 is 38, class 2 is 29 and of class 3 is 24. And there are outliers present in class 2 aand class 3.

```python
m=df.age.isnull()
d = {1: 38, 2: 29, 3: 24}
df.loc[m, 'age'] = df.loc[m, 'pclass'].map(d)
```
